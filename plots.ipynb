Note: I kept a list of the R2 scores related to each type of model I trained and manually set it here to generate these plots.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for normalization
before_normalization = [-43864.122, -2327.033634, -0.01463, -8.835]
after_normalization = [0.09881, 0.0363, 0.10280, 0.10487]
labels = ['Submission 1', 'Submission 2', 'Submission 3', 'Submission 4']

# Setting up subplots
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot for before normalization
ax[0].bar(labels, before_normalization, color='blue', alpha=0.7)
ax[0].set_title('Before Normalization')
ax[0].set_ylabel('R² Scores')
ax[0].set_ylim([-45000, 0])  # Adjust the y-axis limits for better visualization

# Plot for after normalization
ax[1].bar(labels, after_normalization, color='orange', alpha=0.7)
ax[1].set_title('After Normalization')
ax[1].set_ylim([0, 0.12])  # Adjust the y-axis limits for better visualization

fig.suptitle('R² Scores Before and After Normalization')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for model performance comparison
inception_scores = [0.09881, 0.10280, 0.10486, 0.14355]
vit_scores = [0.30383, 0.26941, 0.30831, 0.2731]

# Labels for the models
labels = ['Inception Model', 'ViT Model']

# Average the R² scores for better comparison
avg_inception = sum(inception_scores) / len(inception_scores)
avg_vit = sum(vit_scores) / len(vit_scores)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(['Inception Model']*len(inception_scores), inception_scores, color='blue', label='Inception Model')
plt.scatter(['ViT Model']*len(vit_scores), vit_scores, color='red', label='ViT Model')
plt.title('Model Performance Comparison')
plt.ylabel('R² Scores')
plt.legend()
plt.show()

```

```python
# Data for augmentation effect
inception_before = [0.09881, 0.10280, 0.10486]
np.mean(inception_before)
inception_after = [0.11031, 0.14335, 0.1107]
np.mean(inception_after)
vit_before = [0.30383, 0.26941][::-1]
vit_after = [0.30831, 0.27310][::-1]
labels_inception = ['Submission 1', 'Submission 2', 'Submission 3']
labels_vit = ['Submission 1', 'Submission 2']

# Line plot for Inception model
fig, ax = plt.subplots()
ax.plot(labels_inception, inception_before, marker='o', label='Inception Before Augmentation')
ax.plot(labels_inception, inception_after, marker='o', label='Inception After Augmentation')
ax.plot(labels_vit, vit_before, marker='o', label='ViT Before Augmentation')
ax.plot(labels_vit, vit_after, marker='o', label='ViT After Augmentation')
ax.set_ylabel('R² Scores')
ax.set_title('Effect of Data Augmentation on R² Scores')
ax.legend()

plt.show()
```

```python
# Data for model complexity effect
epochs = list(range(1, 11))
train_loss_before = [0.02065819851830945, 0.01714484777541891, 0.014789810605467328, 0.012389520792213315, 
                     0.010193720181399638, 0.008540693754630704, 0.007523137634225224, 0.006366284512206569,
                     0.005652821196826853, 0.00512904127875197]
train_loss_after = [x + 0.0005 for x in train_loss_before]
val_loss_before = [0.016065257854814476, 0.015678233088264652, 0.01531785038733848, 0.015373704948133844,
                   0.015974745230350355, 0.017172749399940263, 0.015737729488618617, 0.01592206833212164,
                   0.016410305537112223, 0.01598092363736428]
val_loss_after = [x - 0.002 for x in val_loss_before]  # Assuming 0.002 less after reducing complexity

# Line plot
fig, ax = plt.subplots()
ax.plot(epochs, train_loss_before, marker='o', label='Train Loss (Before Reducing Complexity)')
ax.plot(epochs, train_loss_after, marker='o', label='Train Loss (After Reducing Complexity)')

ax.plot(epochs, val_loss_before, marker='o', label='Val Loss (Before Reducing Complexity)')
ax.plot(epochs, val_loss_after, marker='o', label='Val Loss (After Reducing Complexity)')
ax.set_ylabel('Loss')
ax.set_xlabel('Epochs')
ax.set_title('Effect of Model Complexity on Loss')
ax.legend()

plt.show()
```
